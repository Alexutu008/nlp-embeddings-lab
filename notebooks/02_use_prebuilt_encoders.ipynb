```python
import sys
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent
SRC_PATH = PROJECT_ROOT / "src"

if str(SRC_PATH) not in sys.path:
    sys.path.append(str(SRC_PATH))
```

```python
from classical.tfidf_encoder import TfidfEncoder
from static.word2vec_encoder import Word2VecEncoder
from contextual.sentence_encoder import SentenceEncoder
```

```python
texts = [
    "I love natural language processing",
    "Word embeddings are useful for text tasks",
    "Transformers produce contextual representations",
    "Football and sport education are important",
    "Sentence embeddings help semantic search"
]
```

```python
tfidf = TfidfEncoder(max_features=50)
tfidf.fit(texts)
X_tfidf = tfidf.encode(texts)

print(X_tfidf.shape)
print(tfidf.get_feature_names()[:15])
```

```python
w2v = Word2VecEncoder(vector_size=50, sg=1, epochs=50)
w2v.fit(texts)
X_w2v = w2v.encode(texts)

print(X_w2v.shape)
print(w2v.most_similar("embeddings"))
```

```python
sent = SentenceEncoder(model_name="all-MiniLM-L6-v2")
X_sent = sent.encode(texts)

print(X_sent.shape)
```
